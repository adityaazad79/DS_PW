```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['total_bill'].max()
```

```python
df['total_bill'].min()
```

```python
# Distribution Plot (Histogram)
sns.set() # view the grid in the plot
# sns.distplot(df['total_bill'], kde = False) # Will be removed in future. kde = Kernel Density Estimation Function
sns.histplot(df['total_bill'])
```

```python
sns.distplot(df['total_bill'], hist = False) # Plots only kde function
```

```python
sns.scatterplot(x='total_bill', y ='tip', data=df) # (x-axis, y-axis, source_of_data)
```

```python
# Relationship Plot

sns.relplot(x = 'total_bill', y = 'day', data = df)
```

```python
# Bivariate Analysis
# Joint Plot
sns.jointplot(x = 'total_bill', y = 'tip', data=df)
```

```python
# Represent all plots in a single graph

sns.pairplot(df)
```

```python
# Box Plot
# Useful for outliers - If values going beyond a given range
# It helps us with the 5 number summary
# 0th percentile
# 25th percentile
# 50th percentile # Median
# 75th percentile
# 100th percentile

sns.boxplot(x=df['total_bill'])
```

```python
# Categorical Plot

sns.catplot(x = 'day', y = 'total_bill', data = df, kind = 'strip') # kind = 'strip' is default
sns.catplot(x = 'day', y = 'total_bill', data = df, kind = 'box')
sns.catplot(x = 'day', y = 'total_bill', data = df, kind = 'boxen')
```
