```python
import pandas as pd
df = pd.read_csv('/content/Salary_Data (1).csv')
df.head(2)
```

```python
x = df['YearsExperience'].values.reshape(-1, 1)
y = df['Salary']
x
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x, y)
```

```python
# 'm' or 'slope of data'
reg.coef_
```

```python
# 'c' or 'intercept'
reg.intercept_
```

```python
# Predicting the salary for an experience of 3.5 years based on the learning of the given dataset.
reg.predict([[3.5]])
```
