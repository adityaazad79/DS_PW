### Step17. What is the average revenue for the order of item quantity equals to 4?

```python
# df[df['quantity'] == 4]
df4 = df[df['quantity'] == 4]
df4_qty = df4['order_id'].nunique()
df4_t_rev = (df4['item_price'] * df4['quantity']).sum()
df4_avg_rev = df4_t_rev / df4_qty
df4_avg_rev
```
