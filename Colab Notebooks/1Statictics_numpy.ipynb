```python
# It contains arrays as basic object
import numpy as np
#  List is a bit slow and less memory efficient compared to arrays
a = np.array([33, 12, 66, 73, 45, 98, 29, 12])  # Method
print(a)
print(type(a))
print(a.shape)  # Dimension of array

# basic arithmetic operations
print(np.sum(a))

# Central tendency / mean : Average : (xÌ„)
# Sum of total values divided by count
# mean_a = np.sum(a)/len(a)
# print(mean_a)
print(f"The mode value is {np.mean(a)}")


# Median : Arrange the values in an array in ascending order and check the middle value
# Median is alose called --> 50th percentile --> Median(2nd Quartile)
# print(np.sort(a))
np.sort(a)
# print((33+45)/2) # Sum of middle 2 elements divided by 2 in can case even length else just the middle number
print(f"The median value is {np.median(a)}")


# Mode : Most repeated value in a data set
# numpy.mode(array) doesn't exist in numpy library. It's in scipy
print(np.unique(a, return_counts=True)) # np.unique = returns unique elemets in the array and their no of occurence for each position
```

```python
# Take 2 numpy arrays (2 points on coordinate geometry)
# Euclidean Distance on the 2 points to get distance

x1 = np.array([2, 3])
x2 = np.array([5, 7])
dist = ((x1[0] - x2[0])**2 + (x1[1] - x2[1])**2)**(1/2)
print(dist)
```

```python
x1 = np.array([34, 45])
x2 = np.array([29, 41])
# dist = ((x1[0] - x2[0])**2 + (x1[1] - x2[1])**2)**(1/2)
# or
dist1 = np.sqrt(np.sum(np.square(np.subtract(x1,x2))))
print(dist1)
# or
dist2 = np.linalg.norm(x1-x2) # the backend of this function same as above
print(dist2)
```

```python
# pandas library
# Panel data/Panel Data Analysis
# A python library used for working with data sets.
# Explanatory data analysis
# Has 2 main objectives
# Series - 1-D - (1 Row / 1 Column) - vectors in numpy
# Data Frame - 2-D - matrix in numpy
```

```python
d = {'Name' : ['Mr V', 'Mr W', 'Mr X', 'Mr Y', 'Mr Z'],
     'Height' : [123, 121, 118, 110, 129],
     'Weight' : [76, 36, 64, 43, 71]}

import pandas as pd
df=pd.DataFrame(d)
print(df)
print("\nShape : ", df.shape) # Dimension of data dictionary
print("\nDescribe : \n", df.describe()) # Differnt metric attribute of the data
```

```python
# Data visualisation libraries : matplotlib, Seaborn, Plotly
# 1. Line Graph - Join axial points with a line
# 2. Bar Graph - Count of the occurences
# 3. Pie Chart - Undersatandig the composition
# 4. Scatter Plot - Similar to line graph but without the lines. Only the points
# 5. Histogram - Gives the count like a bar graph but will give the values with respect to its frequency.
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([2, 5, 6, 2, 45, 23])
b = np.array([5, 8, 34, 89, 34, 4])
plt.plot(a,b, marker = 'o') # Line plot
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')

plt.show()
```

```python
# Bar Graph

plt.bar(df['Name'], df['Height'], label = 'Height')
plt.bar(df['Name'], df['Weight'], label = 'Weight')
plt.legend()
plt.show()
```

```python
# Pie Chart

cities  = np.array(['Bangluru', 'Chennai', 'Kolkata','Patna', 'Hydrabad', 'Chandigarh'])
sales = np.array([12312,45433,56755,23445,67655,87684])
# plt.pie(sales)
# plt.pie(sales, labels = cities)
plt.pie(sales, labels = cities, autopct = '%.2f') # Composition Graph
# plt.legend()
plt.show()
# plt.pie? # Help
```

```python
# Scatterplot Graph

plt.scatter(df['Height'],df['Weight'])
plt.show()
```

```python
# Histogram - 'hist'

x = np.random.normal(100, 10, 250) # (LOC(The highest value the graph will reach around), Scale, Size(No of values))
plt.hist(x)
plt.show()
```
