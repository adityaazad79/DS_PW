## Merging operation on two dataframes
## Works on columns

```python
data1 = pd.DataFrame({
    'key1': [1, 2, 3, 4, 5],
    'key2': [4, 5, 6, 7, 8],
    'key3': [6, 7, 8, 9, 10],
})
data1
```

```python
data2 = pd.DataFrame(
    {
        'key1': [45, 5, 3, 56, 1],
        'key4': [345, 45, 4343, 1, 41],
        'key5': [534, 234, 335, 45, 2]
    }
)
data2
```

## Inner merge
## works like natural join

```python
pd.merge(data1, data2)
```

## left outer merge
#### Get all from left but common from right table

```python
pd.merge(data1, data2,how='left')
```

## right outer merge
#### Get all from right but common from left table

```python
pd.merge(data1, data2,how='right')
```

## full outer merge
#### Get all from both tables and put null on empty values

```python
pd.merge(data1, data2,how='outer',on="key1")
# Can alos use lefton and righton
```

## Cross Merge
#### map all values of table1 with all values of table2

```python
pd.merge(data1, data2,how='cross')
```

## Join Operation on DataFrames
## Works on rows/index

```python
data1 = pd.DataFrame(
    {
        'key1': [1, 2, 3, 4, 5],
        'key2': [4, 5, 6, 7, 8],
        'key3': [6, 7, 8, 9, 10]
    },
    index=['a', 'b', 'c', 'd', 'e']
)
data1
```

```python
data2 = pd.DataFrame(
    {
        'key11': [45, 5, 3, 56, 1],
        'key22': [345, 45, 4343, 1, 41],
        'key33': [534, 234, 335, 45, 2]
    },
    index=['a', 'b', 'h', 'j', 'j']
)
data2
```

## Natural Join

```python
data1.join(data2)
```

## right outer join

```python
data1.join(data2, how="right")
```

## Inner Join

```python
data1.join(data2,how='inner')
```

## Full Outer Join

```python
data1.join(data2,how='outer')
```

## Cross Join

```python
data1.join(data2,how='cross')
```

## 'apply' function
- Applies a function to the dataframe/series data

- Conversion of fare from euro to rupee

```python
df['Fare_INR'] = df['Fare'].apply(lambda x : x*80)
```

```python
df.head()
```

```python
def euro_inr(x):
    return x*80

df["Fare_INR2"] = df["Fare"].apply(euro_inr)
df.head()
```

```python
# Storing the lenght of Name in a new column for each tuple

df["Len_name"] = df["Name"].apply(len)
```

```python
df.head()
```

```python
# Adding a new column for category based on fare

def cat_fare(x):
    if x < 10:
        return "cheap"
    elif x >= 10 and x < 20:
        return 'mid'
    else:
        return 'high'


df['category_fare'] = df['Fare'].apply(cat_fare)
df.head()
```
