## Decision Tree Classifier Implementation With Post Preprunning


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
```

```python
# df = sns.load_dataset('iris')
```

```python
dataset
```

```python
df2 = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df2
```

## Independent and dependent features


```python
X = df2
y = dataset.target
```

```python
X
```

```python
# Train test Split
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.22, random_state=42)
```

```python
#### We don't need standardization for decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train.head()
```

#### Post-Pruning

```python
tree_classifier = DecisionTreeClassifier()
```

```python
tree_classifier.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier, filled=True)
```

```python
#### Post-prunning with max-depth = 2

tree_classifier=DecisionTreeClassifier(max_depth=2)
tree_classifier.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier, filled=True)
```

```python
## Prediction

y_pred = tree_classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred, y_test)
```

```python
print(score)
```

```python
print(classification_report(y_pred, y_test))
```

```python
# Internal Assignment

from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset
```

```python
df2 = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df2
```

```python
X = df2
y = dataset.target
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.22, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train.head()
```

### Post-Prunning

```python
tree_classifier = DecisionTreeClassifier()
```

```python
tree_classifier.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier, filled=True)
```

#### Post-prunning with max-depth = 2

```python
tree_classifier=DecisionTreeClassifier(max_depth=3)
tree_classifier.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier, filled=True)
```

#### Prediction

```python
y_pred = tree_classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred, y_test)
```

```python
print(score)
```

```python
print(classification_report(y_pred, y_test))
```
