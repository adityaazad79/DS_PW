## Matplotlib
#
#### An object-oriented plotting library.
#### A procedural interface is provided by the companion pyplot module, which may be imported directly

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,200)
x
```

```python
y =np.sin(x)
y
```

```python
plt.plot(x,y)

# Adding decoration : title, xlabel, ylabel

plt.title("sin graph :  wave")
plt.xlabel("Value")
plt.ylabel("sin value")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
```

```python
colors = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
# c = color of data points, s = size of data points, alpha = intensity of color of the data points
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("random data")
plt.show()
```

#### Bar plot

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)
plt.bar(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar plot")
```

```python
y
```

#### Horizontal bar plot

```python
plt.barh(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Horizontal Bar plot")
```

#### x = np.linspace(0,10,200)


```python
# sin graph

x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.figure(figsize=(8, 4)) # Controlling figure size
plt.plot(x, y, '--r')  # graph in red color
plt.title("sin graph :  wave")
plt.xlabel("Value")
plt.ylabel("sin value")
plt.grid() # Adds grid to the background
plt.show()
```

#### Scatter plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x, y, c='#A64D13', s=sizes, alpha=.5, marker='x')
# c = color of the data point, s = size of data points, alpha = Intensity of the data points, marker = shape of the data points, e,g 'v','x', default = 'o'
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

#### Line graph

#### Scatter plot

```python
x = [4, 5, 6, 7, 8, 9]
y = [60, 30, 80, 19, 40, 1]
plt.scatter(x, y)
plt.show()
```

```python
x = [4, 5, 6, 7, 8, 9]
y = [60, 30, 80, 19, 40, 1]
colour = ['red', 'blue', "#4E301D", "#DFC010", "#EC801C", "#DFC6AE"]
# color & color_code
plt.scatter(x, y, c=colour)
plt.show()
```

#### 3d-scatter plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# ax = axis, projection = plane of graph projection
ax.scatter(x, y, z)
```
